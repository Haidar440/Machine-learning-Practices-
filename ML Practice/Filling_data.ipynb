```python
import pandas as pd
```

```python
data = pd.read_csv("Item.csv")
df = pd.DataFrame(data)
df.head()
```

```python
df.fillna(df.mean(),inplace=True)
df.head()
```

```python
# Replacing Missing Value
df["Quantity"].fillna(df["Quantity"].mean(),inplace=True)
df.head()
```

```python
df["Price"].fillna(df["Price"].median(),inplace=True)
df.head()
```

```python
df["Bought"].fillna(df["Bought"].std(),inplace=True)
df.head()
```

```python
df["Forenoon"].fillna(df["Forenoon"].min(),inplace=True)
df.head()
```

```python
df["Afternoon"].fillna(df["Afternoon"].max(),inplace=True)
df.head()
```

```python
df["Item"]=df["Item"].str.upper()
df.head()
```

```python
df["Bought"]=df["Bought"].astype(int)
df.head()
```

```python
print(df[df.duplicated(keep=False)])
```

```python
print(df[df.duplicated()])
```
