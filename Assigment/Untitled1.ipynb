```python
import pandas as pd
import numpy as np
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame(pd.read_csv("laptopData.csv"))
```

```python
df.info()
```

```python
df.isnull().sum()
df.drop(columns="Unnamed: 0",inplace=True)
df.drop(columns="TypeName",inplace=True)
df
```

```python
df["Ram"].fillna(df["Ram"].mode()[0],inplace=True)

df['Ram'] = df['Ram'].str.replace('GB', '', case=False)

# Convert to integer (optional)
df['Ram'] = df['Ram'].astype(int)
```
