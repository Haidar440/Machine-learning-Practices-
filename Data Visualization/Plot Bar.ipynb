```python
import matplotlib.pyplot as plt
x = [10,20,30,40]
y = [20,25,35,55]
plt.plot(x,y)
plt.title("Line Chart")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.show()
```

```python
import seaborn as sns 
x = [10,20,30,40]
y = [20,25,35,55]
sns.lineplot(x=x,y=y)
plt.title("Line Chart")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.show()
```

```python
x = [10,20,30,40]
y = [20,25,35,55]
plt.plot(x,y,color="red",linewidth=2,marker='*',markersize=20,linestyle="--")
plt.title("Line Chart")
plt.ylabel("Y-Axis")
plt.xlabel("X-Axis")
plt.show()
```
